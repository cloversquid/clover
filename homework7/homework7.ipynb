# 2 MatPlotLib

## 2.0 Setup
2.0.1 Import NumPy  
2.0.2 Import Matplotlib  

```python
# 2.0.1 Import NumPy  
# 2.0.2 Import Matplotlib  

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## 2.1 Function
2.1.1 Define function: `make_sine_wave(x, A, w)`  
2.1.2 Return: `A * sin(w * x)`  

```python
def make_sine_wave(x, A, w): # 2.1.1 Define function: `make_sine_wave(x, A, w)`  
    return A * np.sin(w*x) # 2.1.2 Return: `A * sin(w * x)`
```

## 2.2 np.linspace()
2.2.1 Create x array with `np.linspace(0, 2π, 1000)`  

```python
x = np.linspace(0,2*np.pi,1000) # 2.2.1 Create x array with `np.linspace(0, 2π, 1000)`  
```

# 2.3 Create two arrays
2.3.1 Create amplitude array `[0.5, 1, 1.5, 2, 2.5]`  
2.3.2 Create frequency array `[1, 2, 3, 4, 5]` 

```python
amplitude = [0.5, 1, 1.5, 2, 2.5] # 2.3.1 Create amplitude array `[0.5, 1, 1.5, 2, 2.5]` 
```

```python
frequency = [1, 2, 3, 4, 5] # 2.3.2 Create frequency array `[1, 2, 3, 4, 5]`  
```

## Plot Setup
3. Create 3 subplots (figsize = (25, 5)) 

```python
# 2.3.1  

fig, axs = plt.subplots(1, 3, figsize=(25, 5))

line_styles = ['-', '--', '-.', ':', (0, (3, 1, 1, 1))]

payload = list(zip(amplitude, frequency))

for index, value in enumerate(payload): 
    A, w = value
    line = line_styles[index]
    axs[0].plot(x, make_sine_wave(x, A, 1), linestyle=line, label = f"A = {A}, w = 1")   # Step 6
    axs[1].plot(x, make_sine_wave(x, 1, w), linestyle=line, label = f"A = 1, w = {w}")   # Step 6
    axs[2].plot(x, make_sine_wave(x, A, w), linestyle=line, label = f"A = {A}, w = {w}") # Step 6

axs[0].set_title("Varying Amplitude Only")
axs[1].set_title("Varying Frequency Only")
axs[2].set_title("Varying Amplitude and Frequency")

axs[0].set_xlabel("x")
axs[1].set_xlabel("x")
axs[2].set_xlabel("x")

axs[0].set_ylabel("y")
axs[1].set_ylabel("y")
axs[2].set_ylabel("y")

axs[0].legend()
axs[1].legend()
axs[2].legend()

plt.show()

for each in axs:
    print(each)
```
