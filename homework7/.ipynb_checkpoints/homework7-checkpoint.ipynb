# 2 MatPlotLib

## 2.1 Setup
2.1.1 Import NumPy  
2.1.2 Import Matplotlib  

```python
# 2.1.1 Import NumPy  
# 2.1.2 Import Matplotlib  

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## 2.2 Function
2.2.1 Define function: `make_sine_wave(x, A, w)`  
2.2.2 Return: `A * sin(w * x)`  

```python
def make_sine_wave(x, A, w): # 2.2.1 Define function: `make_sine_wave(x, A, w)`  
    return A * np.sin(w*x) # 2.2.2 Return: `A * sin(w * x)`
```

## 2.3 Data Creation
2.3.1 Create x array with `np.linspace(0, 2π, 1000)`  
2.3.2 Create amplitude array `[0.5, 1, 1.5, 2, 2.5]`  
2.3.3 Create frequency array `[1, 2, 3, 4, 5]`  

```python
x = np.linspace(0,2*np.pi,1000) # 2.3.1 Create x array with `np.linspace(0, 2π, 1000)`  
```

```python
amplitude = [0.5, 1, 1.5, 2, 2.5] # 2.3.2 Create amplitude array `[0.5, 1, 1.5, 2, 2.5]` 
```

```python
frequency = [1, 2, 3, 4, 5] # 2.3.3 Create frequency array `[1, 2, 3, 4, 5]`  
```

## 2.4 Plot Setup
2.4.1 Create 3 subplots (figsize = (25, 5)) 

```python
# Step 1
fig, axs = plt.subplots(1, 3, figsize=(25, 5))
```

## 2.5 Looping
2.5.1 Use `zip()` to loop through amplitude and frequency together

```python
for A, w in zip(amplitude, frequency): 
    print(A, w)
# Step 5 Use `zip()` to loop through amplitude and frequency together
bundle = list(zip(amplitude, frequency))
```

## 2.6 Plotting
2.6.1 Plot 1: vary amplitude only (frequency = 1)  
2.6.2 Plot 2: vary frequency only (amplitude = 1)  
2.6.3 Plot 3: vary both amplitude and frequency together  

```python
line_styles = ['-', '--', '-.', ':', (0, (3, 1, 1, 1))]

for i, (A, w) in enumerate(bundle):
    axs[0].plot(x, make_sine_wave(x, A, 1), linestyle=line_styles[i], label=f"A = {A}, w = 1") # 2.6.1 Plot 1: vary amplitude only
    axs[1].plot(x, make_sine_wave(x, 1, w), linestyle=line_styles[i], label=f"A = 1, w = {w}") # 2.6.2 Plot 2: vary frequency only
    axs[2].plot(x, make_sine_wave(x, A, w), linestyle=line_styles[i], label=f"A = {A}, w = {w}") # 2.6.3 Plot 3: vary both amplitude and frequency together
```

## 2.7 Formatting
2.7.1 Add titles  
2.7.2 Add axis labels  
2.7.3 Add legends  

```python
axs[0].set_title("Varying Amplitude Only") # 2.7.1 Add titles
axs[1].set_title("Varying Frequency Only") # 2.7.1 Add titles
axs[2].set_title("Varying Amplitude and Frequency") # 2.7.1 Add titles

axs[0].set_xlabel("x") # 2.7.2 Add axis labels
axs[1].set_xlabel("x") # 2.7.2 Add axis labels
axs[2].set_xlabel("x") # 2.7.2 Add axis labels

axs[0].set_ylabel("y") # 2.7.2 Add axis labels
axs[1].set_ylabel("y") # 2.7.2 Add axis labels
axs[2].set_ylabel("y") # 2.7.2 Add axis labels

axs[0].legend() # 2.7.3 Add legends
axs[1].legend() # 2.7.3 Add legends
axs[2].legend() # 2.7.3 Add legends
```

## 2.8 Display
2.8.1 Call `plt.show()`  

```python
plt.show(block=True) # 2.8.1 Call `plt.show()`
```
